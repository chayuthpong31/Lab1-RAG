## Overview

Retrieval Augmented Generation (RAG) improves Large Language Models (LLMs) by allowing them to access and process external information sources during generation. This ensures the model's responses are grounded in factual data and avoids hallucinations.

A common problem with LLMs is that they don't understand private knowledge, that
is, your organization's data. With RAG Engine, you can enrich the
LLM context with additional private information, because the model can reduce
hallucinations and answer questions more accurately.

By combining additional knowledge sources with the existing knowledge that LLMs
have, a better context is provided. The improved context along with the query
enhances the quality of the LLM's response.

The following concepts are key to understanding Vertex AI RAG Engine. These concepts are listed in the order of the
retrieval-augmented generation (RAG) process.

1. **Data ingestion**: Intake data from different data sources. For example,
  local files, Google Cloud Storage, and Google Drive.

1. **Data transformation**: Conversion of the data in preparation for indexing. For example, data is split into chunks.

1. **Embedding**: Numerical representations of words or pieces of text. These numbers capture the
   semantic meaning and context of the text. Similar or related words or text
   tend to have similar embeddings, which means they are closer together in the
   high-dimensional vector space.

1. **Data indexing**: RAG Engine creates an index called a corpus.
   The index structures the knowledge base so it's optimized for searching. For
   example, the index is like a detailed table of contents for a massive
   reference book.

1. **Retrieval**: When a user asks a question or provides a prompt, the retrieval
  component in RAG Engine searches through its knowledge
  base to find information that is relevant to the query.

1. **Generation**: The retrieved information becomes the context added to the
  original user query as a guide for the generative AI model to generate
  factually grounded and relevant responses.

For more information, refer to the public documentation for [Vertex AI RAG Engine](https://cloud.google.com/vertex-ai/generative-ai/docs/rag-overview).

### Generate Content with Gemini using RAG Retrieval Tool

```
MODEL_ID = "gemini-2.5-flash"
```

```
def ask_rag(query, MODEL_ID, rag_retrieval_tool):
  # if query not in doc return this
  system_prompt = """Answer ONLY using the provided context.
    If the answer is not in the documents, say: 'I cannot find this in our records.'"""

  response = client.models.generate_content( # Call the generate_content method from the client's models.
      model=MODEL_ID, # Specify the generative model to use (e.g., 'gemini-2.5-pro').
      contents=f"{system_prompt}\n\nUser Query: {query}", # An example of a commented-out content query.
      config=GenerateContentConfig(tools=[rag_retrieval_tool]), # Configure content generation, including the RAG retrieval tool.
  )

  display(Markdown(f"**Q:** {query}"))
  display(Markdown(f"**A:** {response.text}"))
  print("-" * 30) # Display the generated response as Markdown.
```

### Chat with the RAG-enabled Model

This loop allows for continuous conversation with the generative model, leveraging the RAG retrieval tool. Type `end` to stop the chat.

```
import time

test_cases = {
    "Fact-Check": [
        "How many credits are required for the Bachelor International Program?",
        "How many credits are required for the Bachelor Regular Program?",
        "How many credits are required for the Master's Degree program?",
        "How many credits are required for the Doctor of Philosophy Programme?",
        "In which year of the Bachelor program is the internship required?",
        "How long is the internship duration for the Bachelor program?",
        "What CPE 112 learning about?",
        "What CPE 314 learning about?",
        "What is the total tuition fee for the Master Program?",
        "What CPE 624 learning about",
    ],
    "Out-of-Bounds": [
        "What is the capital of Thailand?",
        "How to cook pad thai?",
        "What is the largest country in the world?",
        "Who invented alternating current?"
    ],
    "Ambiguity (3)": [
        "How do the electives in the International Program differ from the Regular Program, and are there any courses with the same name but different course codes?",
        "Compare the admission requirements between the Master's and Doctoral programs; what are the similarities and differences?",
        "Across all levels of education (Bachelor, Master, and PhD), in which year is each 'Artificial Intelligence' related course offered in their respective programs?"
    ]
}

for category, queries in test_cases.items():
    print(f"\n=== Testing {category} ===")
    for q in queries:
        try:
            ask_rag(q, MODEL_ID, rag_retrieval_tool)
            print("Waiting for quota reset...")
            time.sleep(10)
        except Exception as e:
            if "429" in str(e):
                print("Rate limit hit! Resting for 60 seconds...")
                time.sleep(60)
            else:
                print(f"Error processing '{q}': {e}")
```

### Chunk Data

```
def chunk_data():
  #reading from a public bucket.  This bucket is hosted by someone else so it's free.
  INPUT_GCS_BUCKET = (
      "gs://kmutt-rag-lab-1/RAG_LAP1/cpe_bachelor_reg.pdf"
  )
  print("Start reading and chunking data in gcs bucket ", INPUT_GCS_BUCKET, " ...")

  response = rag.import_files(
      corpus_name=rag_corpus.name,
      paths=[INPUT_GCS_BUCKET],
      # Optional
      transformation_config=rag.TransformationConfig(
          chunking_config=rag.ChunkingConfig(chunk_size=1024, chunk_overlap=100)
      ),
      max_embedding_requests_per_min=900,  # Optional
  )
  print("Chunking of data in Public GCS Bucket done.")
```

### Create rag tool

```
def create_rag_retrieval_tool(rag_corpus):
  # Create a tool for the RAG Corpus
  rag_retrieval_tool = Tool(
      retrieval=Retrieval(
          vertex_rag_store=VertexRagStore(
              rag_corpora=[rag_corpus.name],
              similarity_top_k=10
          )
      )
  )

  return rag_retrieval_tool
```

### Find Ground Truth ID

```
def ground_truth_id(query, MODEL_ID, rag_retrieval_tool):
    system_prompt = """Answer ONLY using the provided context.
    If the answer is not in the documents, say: 'I cannot find this in our records.'"""

    response = client.models.generate_content(
        model=MODEL_ID,
        contents=f"{system_prompt}\n\nUser Query: {query}",
        config=GenerateContentConfig(tools=[rag_retrieval_tool]),
    )

    gt_ids = []
    print(f"\n[QUERY]: {query}")

    if response.candidates[0].grounding_metadata.grounding_chunks:
        chunks = response.candidates[0].grounding_metadata.grounding_chunks
        print(f"--- Ground Truth Chunks Found: {len(chunks)} ---")

        for i, chunk in enumerate(chunks):
            content_text = chunk.retrieved_context.text
            # สร้าง ID สำหรับใช้อ้างอิง (หรือใช้ content_text เลยเพื่อความแม่นยำในการหา recall)
            unique_chunk_id = content_text
            gt_ids.append(unique_chunk_id)

            # (ข้อ a) แสดงเลข Chunk และเนื้อหาที่เกี่ยวข้อง
            print(f"Ground Truth Chunk #{i+1}:")
            print(f"Content: {content_text[:300]}...") # พิมพ์ 300 ตัวอักษรแรกที่เกี่ยวข้อง
            print("-" * 20)
    else:
        print("AI: 'I cannot find this in our records.'")

    return gt_ids
```

```
def show_top_k_results(query, rag_retrieval_tool):
    # ดึงเฉพาะผลลัพธ์การ Search (Retrieval) โดยไม่ผ่าน LLM
    # หมายเหตุ: ใน Vertex AI SDK คุณสามารถเรียกจาก rag_retrieval_tool ได้โดยตรง
    response = client.models.generate_content(
        model="gemini-2.5-flash", # ใช้รุ่นไหนก็ได้เพื่อดึง tool
        contents=query,
        config=GenerateContentConfig(tools=[rag_retrieval_tool]),
    )

    print(f"\n--- Top 10 Retrieved Chunks for this Embedding ---")
    if response.candidates[0].grounding_metadata.grounding_chunks:
        chunks = response.candidates[0].grounding_metadata.grounding_chunks[:10]
        for i, chunk in enumerate(chunks):
            text = chunk.retrieved_context.text
            # ตัดมาเฉพาะ 2-3 บรรทัดแรก
            lines = text.split('\n')[:3]
            snippet = "\n".join(lines)

            print(f"Rank {i+1}:")
            print(f"{snippet}...")
            print("-" * 10)
```

### Calculate Recall@10

```
def recall_at_10(query, MODEL_ID, rag_retrieval_tool, gt_ids):
    response = client.models.generate_content(
        model=MODEL_ID,
        contents=query,
        config=GenerateContentConfig(tools=[rag_retrieval_tool]),
    )

    retrieved_texts = []
    if response.candidates[0].grounding_metadata.grounding_chunks:
        # ดึงมาเฉพาะ 10 อันดับแรก
        chunks = response.candidates[0].grounding_metadata.grounding_chunks[:10]
        for c in chunks:
            # ทำความสะอาดเล็กน้อย: ตัดช่องว่างและขึ้นบรรทัดใหม่
            clean_text = " ".join(c.retrieved_context.text.split())
            retrieved_texts.append(clean_text)

    if not gt_ids:
        return 0.0

    correct_retrieved_count = 0

    # วนลูปเช็ค Ground Truth แต่ละอัน
    for gt in gt_ids:
        # ทำความสะอาด GT ให้เหมือนกับที่ทำกับ Retrieval
        clean_gt = " ".join(gt.split())

        # ใช้ Substring match: เช็คว่า GT บางส่วน (เช่น 50 ตัวแรก)
        # ปรากฏอยู่ในรายการที่ดึงมา (retrieved_texts) อันไหนบ้างหรือไม่
        match_found = any(clean_gt[:50] in rt for rt in retrieved_texts)

        if match_found:
            correct_retrieved_count += 1

    recall_score = correct_retrieved_count / len(gt_ids)
    return recall_score
```

### Create 002 Corpus RAG

```
# Currently supports Google first-party embedding models
# fmt: off
# EMBEDDING_MODEL = "publishers/google/models/text-embedding-005"  # @param {type:"string", isTemplate: true}
EMBEDDING_MODEL = "publishers/google/models/text-multilingual-embedding-002"  # @param {type:"string", isTemplate: true}
# fmt: on

rag_corpus = rag.create_corpus(
    display_name="lab4.2_corpus_002",
    backend_config=rag.RagVectorDbConfig(
        rag_embedding_model_config=rag.RagEmbeddingModelConfig(
            vertex_prediction_endpoint=rag.VertexPredictionEndpoint(
                publisher_model=EMBEDDING_MODEL
            )
        )
    ),
)
```

```
rag.list_corpora()
```

```

import time

chunk_data()
rag_retrieval_tool = create_rag_retrieval_tool(rag_corpus)

llm_model = [
    "gemini-2.5-flash"
]

data = [
    {"query":"หลักสูตรวิศวกรรมคอมพิวเตอร์ภาคปกติเรียนกี่หน่วยกิต?", "recall": 0},
    {"query":"วิชา CPE 402 มีวิชาบังคับก่อนไหม?", "recall": 0},
    {"query":"วิชา Database เรียนชั้นปีที่เท่าไหร่?", "recall": 0},
    {"query":"วิชา CPE 333 เรียนเกี่ยวกับอะไร?", "recall": 0},
    {"query":"ชั้นปีที่ 3 เรียนวิชาอะไรบ้าง?", "recall": 0}
]

gt_results = {}

print("\n" + "="*50)
print("--- Start Finding Ground Truth (Step a) ---")
print("="*50)

for i, item in enumerate(data):
    success = False
    while not success:
        try:
            # ใช้ฟังก์ชันใหม่ที่สั่ง Print ข้อมูลออกมาด้วย
            gt_ids = ground_truth_id(item["query"], "gemini-2.5-flash", rag_retrieval_tool)
            gt_results[i] = gt_ids
            print(f"Done for Q{i+1}")
            success = True
            time.sleep(10)
        except Exception as e:
            if "429" in str(e):
                print("Quota Limit sleep 30s...")
                time.sleep(30)
            else:
                print(f"Error : {e}")
                break

results = []

print("\n" + "="*50)
print("--- Testing Embedding 002 Evaluation (Step b) ---")
print("="*50)

for llm_name in llm_model:
    print(f"\n[Model Testing]: {llm_name}")
    scores = []

    for i in range(len(data)):
        print(f"\n>>> Question {i+1}: {data[i]['query']}")

        # แสดง Top 10 Retrieval ตามโจทย์ข้อ b
        show_top_k_results(data[i]["query"], rag_retrieval_tool)

        success = False
        while not success:
            try:
                # คำนวณ Recall ตามโค้ดเดิมของคุณ
                score = recall_at_10(data[i]["query"], llm_name, rag_retrieval_tool, gt_results[i])
                scores.append(score)
                print(f"✅ Q{i+1} Recall Score: {score}")
                success = True
                time.sleep(10)
            except Exception as e:
                if "429" in str(e):
                    time.sleep(30)
                else:
                    print(f"Error: {e}")
                    success = True
                    scores.append(0)

    overall = sum(scores) / len(scores)
    row = ['002', llm_name]
    results.append(row + scores + [overall])

# สร้าง DataFrame สรุป
df1 = pd.DataFrame(results, columns=["Embedding", "Model", "Q1", "Q2", "Q3", "Q4", "Q5", "Overall"])
display(df1)
```
